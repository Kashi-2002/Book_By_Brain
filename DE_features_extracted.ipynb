***The data used here is the preprocessed seed dataset that was in the form of npz file which contained array corresponding to each person.It was then transformed to a dataframe in which each column specified a paticular feature.*** 

```python
import numpy as np
import pickle
import os
import pandas as pd

# check the version of these modules
print(np.__version__)
print(pickle.format_version)
```

```python
os.chdir('D:\EEG to emotion\EEG_DE_features')
```

```python
os.listdir()
```

```python
final_data=np.zeros((16*1823,310))
final_label=np.zeros((16*1823,1))
```

```python
m=0
for i in range(16):
    path=os.listdir()[i]
    data_npz = np.load(path)
    data = pickle.loads(data_npz['data'])
    label = pickle.loads(data_npz['label'])
    for j in range(45):
        for l in range(data[j].shape[0]):
            for k in range(310):
                final_data[m][k]=data[j][l][k]
                final_label[m]=label[j][l]
            
            m=m+1
```

```python
df=np.concatenate((final_data,final_label),axis=1)
```

```python
col_names=[]
for i in range(310):
    col_names.append('feature'+str(i))
    
col_names.append('label')
```

```python
seed_2=pd.DataFrame(df,columns=col_names)
seed_2
```

```python
seed_2.to_csv('seed_2.csv')
```
